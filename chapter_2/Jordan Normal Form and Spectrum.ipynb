```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
n=4

# A unifornmly distributed, positive, row stochastic matrix.
m1 = np.random.uniform(0, 1,(n,n))
m1 = m1 / m1.sum(axis=1, keepdims=1)

# A unifornmly distributed, positive, row sub-stochastic matrix.
m2 = np.random.uniform(0, 1,(n,n))
m2 = m2/(m2.sum(axis=1, keepdims=1) + m2[:,0])
```

```python
# Scatter plot of eigs in complex plane overalyed over unit circle
def plot_spectrum(M):
    eigvals = np.linalg.eig(M)[0]
    reals = [i.real for i in eigvals]
    imags = [i.imag for i in eigvals]
    
    fig, ax = plt.subplots(figsize=(4,4))
    plt.scatter(reals, imags)
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    
    circle=patches.Circle((0,0),radius=1,alpha=.1)
    ax.add_patch(circle)
    return eigvals, fig, ax
```

```python
plot_spectrum(m1)
plot_spectrum(m2)
```

```python
# Scatter plot of eigenvalues overlayed with gergorin disks and unit circle
def plot_gersgorin_disks(M):
    eigvals, fig, ax = plot_spectrum(M)
    
    row_sums = M.sum(axis=1)
    for i in range(M.shape[0]):
        ax.add_patch(patches.Circle((M[i,i],0),radius=row_sums[i]-M[i,i],
                                    alpha=.6/M.shape[0]))
```

```python

plot_gersgorin_disks(m1)
plot_gersgorin_disks(m2)
```
